```python
hor = runs_df.horse_id.unique().tolist()
```

```python
results_dict = {horse: None for horse in hor}
finish_time_dict = {horse: None for horse in hor}
length_behind_dict = {horse: None for horse in hor}
```

```python
len(results_dict)
```

```python
results = []
finish_time = []
length_behind = []
placed = []

# calculate some new columns
for name, row in runs_df.iterrows():
    p = 1 if row['result'] < 4 else 0
    placed.append(p)
    
    results.append(results_dict[row['horse_id']])
    finish_time.append(finish_time_dict[row['horse_id']])
    length_behind.append(length_behind_dict[row['horse_id']])
    
    results_dict[row["horse_id"]] = row["result"]
    finish_time_dict[row["horse_id"]] = row["finish_time"]
    length_behind_dict[row["horse_id"]] = row["lengths_behind"]
    
# add the new columns
runs_df["prev_result"] = results
runs_df["prev_finish_time"] = finish_time
runs_df["prev_lengths_behind"] = length_behind
runs_df["placed"] = placed
runs_df.shape
```
